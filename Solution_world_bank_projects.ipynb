```python
# import packages needed
import json
from pandas.io.json import json_normalize
import pandas as pd
```

```python
# load json as string
x =json.load((open('C:/Users/darsa/Downloads/data_wrangling_json/data_wrangling_json/data/world_bank_projects.json')))
```

```python
# use normalization to create tables from nested element
# reset index of the dataframe
exercise_1=json_normalize(x,'_id',['countryshortname','project_name'])
```

```python
# top 10 Count of the number of time country appears in observation with projects in descending order.
# Multiple Solution for exercise 1 

# using dataframe's value_counts.
solution_1 = exercise_1.countryshortname.value_counts(dropna= True).head(10)

#print the results.
solution_1
```

```python
# Or 
# using dataframe's groupby
solution_1 = exercise_1.groupby('countryshortname')['countryshortname'].count().sort_values( ascending=False).head(10)

#print the results.
solution_1
```

```python
# use normalization to create tables from nested element
# reset index of the dataframe

exercise_2 = json_normalize(x,'mjtheme_namecode',['countryshortname','project_name'])
```

```python
# top 10 Count of the number of time project themes appears in observation with country in descending order.
# Multiple Solution for exercise 2

# using dataframe's groupby
solution_2 = exercise_2.groupby('name')['countryshortname'].count().sort_values(ascending=False).head(10)

#print the results.
solution_2
```

```python
# Or
# using dataframe's value_counts.
solution_2 = exercise_2.name.value_counts(dropna= False).head(10)

#print the results.
solution_2
```

```python
# subset dataframe for creating unique key, value pair for name and code.
exercise_3 = exercise_2.loc[:,['name','code']]
```

```python
# Clean dataframe by removing duplicate and empty name observations
unq_rec = exercise_3[exercise_3.name !=''].drop_duplicates()
```

```python
# create and set index of the dataframe to 'Code' column of dataframe unq_rec
unq_rec.set_index('code',inplace=True)
```

```python
#make a copy of dataframe from exercise_2 data
solution_3 = exercise_2.copy()

#Update solution_3['name'] column to fill the empty value with value of the dictionary by mapping index of dataframe to dictionry key
solution_3['name'].update(exercise_3['code'].map(unq_rec['name'].to_dict()))
```
